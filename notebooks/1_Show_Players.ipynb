# TWBC Database Analysis

## 1. Show table Players

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('../data/processed/twbc.db')

# Query the players table
df_players = pd.read_sql("SELECT * FROM players", conn)

# Display the results
df_players
```

## 2. Custom

```python
df_tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%'", conn)
df_tables
```

```python
pd.set_option('display.max_rows', None)
pd.read_sql("""
    SELECT * FROM players
    LIMIT 20
    """, conn)
```

```python
pd.read_sql("""
    select 
        full_name   "Full Name",
        t.name      "Team Name"
    from 
        players p
        left join teams t on p.team_id = t.team_id
    order by
        t.name
    """, conn)
```
