## 1. Show table Players

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('../data/processed/twbc.db')

# Query the players table
df_players = pd.read_sql("SELECT * FROM players", conn)

# Display the results
df_players
```

## 2. Custom

```python
df_leaderboard = pd.read_sql("""
    SELECT
        RANK() OVER (ORDER BY total_score DESC, efficiency DESC) AS "Place",
        player_name AS "Player",
        player_nick AS "WBC nick",
        team AS "Team",
        total_score AS "Score",
        wins AS "Wins",
        draws AS "Draws",
        losses AS "Losses",
        PRINTF('%.1f %%', efficiency * 100) AS "Efficiency",
        total_games AS "Games"
    FROM v_player_wld
    ORDER BY "Place";
""", conn)

# Display the dataframe
df_leaderboard
```
